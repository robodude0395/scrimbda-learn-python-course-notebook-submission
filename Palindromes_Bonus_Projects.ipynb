```python
#Return largest palindrome of two 3 digits number products (range is defined)

import time
```

```python
#Palindrome function
def is_palindrome(val):
    val = str(val)
    return val == val[::-1]
```

```python
#Constants
max_val = 999
min_val = 100
iterations = 0
min_num2_val = min_val

#Palindrome lists (including debug)
palindrome_list = []
debug_list = []
```

```python
# Brute force approach
initial_time = time.time()

for num1 in range(min_val, max_val):
    for num2 in range(min_val, max_val):
        iterations += 1
        product = num1*num2
        if(is_palindrome(product)):
            palindrome_list.append(product)
            debug_list.append([num1, num2, product])

print(f"Final palindrome list: {palindrome_list}")
print(f"lenght: {len(palindrome_list)}")
print(f"Debug list: {debug_list}")
print(f"Done in {time.time() - initial_time} seconds")
print(f"Iterations: {iterations}")
```

```python
#Optimized approach
iterations = 0
initial_time = time.time()
palindrome_list = []
debug_list = []

for num1 in range(max_val, min_val, -1):
    if(num1 < min_val):
        break
    for num2 in range(max_val, min_num2_val, -1):
        iterations += 1
        product = num1*num2
        if (is_palindrome(product)):
            palindrome_list.append(product)
            min_val = max((product)/max_val, min_val)
            min_num2_val = num2
            debug_list.append([num1, num2, product])
        if(num1 == num2):
            break

print(f"Final palindrome list: {palindrome_list}, lenght: {len(palindrome_list)}")
print(f"Lenght: {len(palindrome_list)}")
print(f"Debug list: {debug_list}")
print(f"Done in {time.time() - initial_time} seconds")
print(f"Iterations: {iterations}")
```

```python
#Further optimized approach
iterations = 0
initial_time = time.time()
palindrome_list = []
debug_list = []

#CODE
nums = ("9","8","7","6","5","4","3","2","1", "0")

#Iterate through all permutations in a different manner
for digit1 in range(10):
    for digit2 in range(10):
        for digit3 in range(10):
            pal_str = nums[digit1] + nums[digit2] + nums[digit3] + nums[digit3] + nums[digit2] + nums[digit1]
            palindrome = int(pal_str)
            
            min_val = int(palindrome/999) + 1
            max_val = int(palindrome**0.5) + 1
            for digit in range(min_val, max_val):
                iterations += 1
                if(palindrome % digit == 0):
                    palindrome_list.append(palindrome)
                    debug_list.append([digit, palindrome/digit, palindrome])
                    break

print(
    f"Final palindrome list: {palindrome_list}")
print(f"Lenght: {len(palindrome_list)}")
print(f"Debug list: {debug_list}")
print(f"Done in {time.time() - initial_time} seconds")
print(f"Iterations: {iterations}")
```
