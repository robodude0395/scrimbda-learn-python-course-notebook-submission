```python
# ü™ê Mars Mission Planner ‚Äî Challenge Steps
#
# 1. Write a function to calculate how long it takes
#    to reach Mars at a given speed.
#    - Mars' average distance is 225,000,000 km.
#    - Use the formula: time = distance √∑ speed
#    - Round the result to the nearest hour.
#
# 2. Write another function to figure out the total fuel cost
#    for a mission.
#    - Formula: total cost = distance √ó fuel usage √ó price per liter
#
# 3. Test your functions with the provided mission data:
#    - Pathfinder: 40,000 km/h
#    - Perseverance: 75,000 km/h
#    - Starship: 120,000 km/h
#    - Each mission travels 225 million km,
#      burns 0.3 liters/km, and fuel costs $1.80/L.
#
# 4. For each mission, print a report showing:
#    - Mission name
#    - Travel time (hours)
#    - Total fuel cost

# Mission data
mission_1_speed = 40000  # Pathfinder
mission_2_speed = 75000  # Perseverance
mission_3_speed = 120000 # Starship

mars_distance = 225_000_000  # in kilometers
fuel_rate = 0.3              # liters per kilometer
fuel_price = 1.8             # dollars per liter

def ETA(distance_km=None, speed_kmh=None):
    time_hours = -1
    if(distance_km != None and speed_kmh != None):
        time_hours = distance_km/speed_kmh

    return round(time_hours)

def fuel_cost(distance_km, fuel_usage, price_per_liter):
    total_fuel = distance_km * fuel_usage
    cost = total_fuel * price_per_liter
    return cost


print("===== Mars Mission Planner =====\n")
print("Pathfinder Mission")
print(f"ETA (hours): {ETA(mars_distance, mission_1_speed)}")
print(f"Fuel cost: {fuel_cost(mars_distance, fuel_rate, fuel_price)}")
print()

print("Perseverance Mission")
print(f"ETA (hours): {ETA(mars_distance, mission_2_speed)}")
print(f"Fuel cost: {fuel_cost(mars_distance, fuel_rate, fuel_price)}")
print()

print("Starship Mission")
print(f"ETA (hours): {ETA(mars_distance, mission_3_speed)}")
print(f"Fuel cost: {fuel_cost(mars_distance, fuel_rate, fuel_price)}")
print()
```

```python
# ‚òï Coffee Order Queue Challenge
# 1. Set up two variables: one for total price, one for drink count
# 2. Start a while True loop
# 3. Ask for the customer's name
# 4. If the name is "done", break the loop
# 5. Ask for their drink order
# 6. If it's "latte", add 3.50 to total and +1 to drink count
#    If it's "americano", add 3.00 to total and +1 to drink count
#    If it's "espresso", add 2.50 to total and +1 to drink count
# 7. If it's not one of those drinks, print a warning and continue
# 8. After the loop, print total number of drinks and total price

drinks_menu = {"latte" : 3.50,
          "americano": 3.0,
          "espresso": 2.5}

order_summary = {"total_price": 0,
                 "total_drink_count": 0}

while True:
    order = input("What  will you order today? ")
    if(order == "done"):
        break

    if(order in drinks_menu.keys()):
        print(f"{order} coming right up!")
        order_summary["total_price"] += drinks_menu[order]
        order_summary["total_drink_count"] += 1

    else:
        print("Not on the menu")

print("---ORDER SUMMARY---")
print(f"Total drinks count: {order_summary['total_drink_count']}")
print(f"Total price: {order_summary['total_price']}")
```

```python
# üì± Phone Number Formatter
#
# 1. Ask the user to enter a U.S. phone number in **any format**.
# 2. Use .strip() to remove any leading/trailing spaces.
# 3. Replace common separators (-, (, ), .) with spaces.
# 4. Use .split() to break into chunks, then .join() to merge the digits.
# 5. Check if the cleaned number has **exactly 10 digits**.
# 6. If yes, format it like this: (123) 456-7890
# 7. If not, print an error message: "Please enter exactly 10 digits."

while True:
    number = input("Enter US phone number: ")

    if(number == "exit"):
        break

    number = number.strip()
    number = number.replace("-", " ")
    number = number.replace(",", " ")
    number = number.replace(".", " ")

    number_arr = number.split(" ")

    num_cleaned = "".join(number_arr)

    print(num_cleaned)

    num_lenght = len(num_cleaned)

    if(num_lenght != 10):
        print("Please enter exactly 10 digits")
        continue
    
    if(num_cleaned.isdigit() == False):
        print("Number can only contain numerical digits")
        continue
    
    print(f"({num_cleaned[:3]}) {num_cleaned[3:6]}-{num_cleaned[6:]}")
```

```python
# üõÇ Access Control Scanner Challenge
#
# 1. Create a set of revoked badge numbers.
# 2. Create two empty lists: "approved" and "denied".
# 3. Start a loop to collect visitor info:
#    - Ask for the visitor's name (or type "done" to finish).
#    - If the name is "done", exit the loop.
#    - Otherwise, ask for their badge number.
#    - Check if the badge is revoked:
#        ‚Ä¢ If revoked: add the name to "denied" and display "ACCESS DENIED".
#        ‚Ä¢ If not: oadd the name to "approved" and display "ACCESS GRANTED".
# 4. Print the final "Access Summary" for "‚úÖ Approved Visitors" & "‚õîÔ∏è Denied Visitors":
#    - Sort both lists alphabetically.
#    - Display the total number of approved and denied visitors.

revoked = {"Z333","A424","B676", "C696", "D999"}
approved = []
denied = []

while True:
    name = input("Enter your name: ")

    #Break out of the loop if name is "done" or name input is blank
    if(name.lower() == "done" or name == ""):
        break

    badge_num = input("Enter badge num: ").strip().upper()
    
    #CHECKS TO SEE IF BADGE NUMBER IS VALID
    if(badge_num[0].isdigit()):
        print("First character must be a letter")
        continue
    
    if(badge_num[1::].isdigit() == False):
        print("Remaining characters after the first on must only be numbers")
        continue

    if(len(badge_num) != 4):
        print("Badge number must only contain 4 characters")
        continue


    if(badge_num in revoked):
        print("ACCESS DENIED")
        if(name not in denied):
            denied.append(name)
    
    else:
        print("ACCESS GRANTED")
        if(name not in approved):
            approved.append(name)

denied.sort()
approved.sort()

print()
print("Access Summary")
print("‚úÖ Approved visitors")
if(len(approved) == 0): print("---No approved visitors---")
else: print(f"Total approved people: {len(approved)}")
for approved_badge in approved:
    print(approved_badge.title())

print()

print("‚õîÔ∏è Denied Visitors")
if(len(denied) == 0): print("---No denied visitors---") 
else: print(f"Total denied people: {len(denied)}")
for denied_badge in denied:
    print(denied_badge.title())
```

```python
# ‚òïÔ∏è Loyalty Points Engine Challenge
#
# RULES:
# ‚Ä¢ Each whole dollar spent earns 3 points
# ‚Ä¢ Tiers:
#     < 100 pts   ‚Üí  Bronze
#     100-499 pts ‚Üí Silver
#     ‚â• 500 pts   ‚Üí  Gold
#
# STEPS:
# 1. Define earn_points(price) ‚Üí returns points for one purchase
# 2. Define tier_label(points) ‚Üí returns "Bronze" / "Silver" / "Gold"
# 3. Given the hard-coded list `purchases`,
#    loop through it, call earn_points() for each amount,
#    and add the result to total_points.
# 4. After the loop, call tier_label(total_points)
# 5. Print 'Loyalty Summary':
#       ‚Ä¢ Total dollars spent
#       ‚Ä¢ Total points earned
#       ‚Ä¢ Final tier

def earn_points(price):
    points = price*3
    return points

def tier_label(points):
    tier = "bronze"

    if(points >= 100 and points <=499):
        tier = "silver"
    
    elif(points >= 500):
        tier = "gold"

    return tier


# Purchase history (e.g., 3.75, 7.20, etc.)
purchases = [3.75, 7.20, 70.1044, 1.05, 90.95, 10.95]

total_dollars = 0
total_points = 0

for purchase in purchases:
    whole_purchase = int(purchase)
    purchase_points = earn_points(whole_purchase)
    print(f"Purchase: ¬£{purchase}")
    print(f"Points earned: {purchase_points}")
    print()
    total_dollars += purchase
    total_points += purchase_points

print("=====Loyalty Summary=====")
print(f"Total spent: ¬£{total_dollars:.2f}")
print(f"Total points earned: {int(total_points)}")
print(f"Final tier: {tier_label(total_points).title()}")
```

```python
# üêæ Dog Bus Tracker ‚Äî Challenge Steps
#
# 1. Start with a bus dictionary holding current passengers.
#    - Each seat number (1, 2, 3, ...) is a key
#    - Each value is another dictionary with each pet's:
#        ‚Ä¢ name
#        ‚Ä¢ breed
#        ‚Ä¢ pickup time
#        ‚Ä¢ dropoff time
#
# 2. Print a starting roster showing each pet‚Äôs seat, name, and pickup time.
#
# 3. Add one new pet if there‚Äôs room on the bus.  
#    - Use MAX_SEATS to limit capacity.   
#    - Dynamically assign the next seat number.  
#    - Print the updated roster showing all pets after pickup.  
#
# 4. Ask which pet leaves early.  
#    - Remove that pet from the bus.  
#    - Print a message saying they‚Äôve headed home.  
#
# 5. Print a final roster listing the remaining pets and their dropoff times.  

import random

def random_time(start=7, end=19):
    """Generate a random hour (HH:00) between given hours."""
    hour = random.randint(start, end)
    return f"{hour:02d}:00"

def return_random_bus_roster():
    names = [
        "Sir Barks A Lot", "Lady Woofsington", "Mr. Snuggles", "Duchess Paws",
        "Captain Wiggles", "Princess Fluffy", "Bark Twain", "Droolius Caesar",
        "Sir Waggington", "Count Drool", "Fur-dinand", "Bark Ruffalo",
        "Chewbacca", "Sandy Paws", "Paw Newman", "Dogtor Strange",
        "Santa Paws", "Pupperoni", "Paw Revere", "Sherlock Bones"
    ]

    breeds = [
        "Terrier", "Labrador", "Poodle", "Bulldog", "Beagle",
        "German Shepherd", "Golden Retriever", "Boxer", "Dachshund", "Corgi"
    ]

    dogs = {}

    skip = [1, 5, 6, 8]

    for i in range(0, len(names)):
        if(i in skip):
            continue
        dogs[i] = {
            "name": names[i],
            "breed": random.choice(breeds),
            "pickup_time": random_time(7, 11),   # morning pickup
            "dropoff_time": random_time(15, 20) # afternoon/evening dropoff
        }

    return dogs

MAX_SEATS = 20
bus = return_random_bus_roster()
bus_seats = list(range(0, MAX_SEATS))

for (seat, dog) in sorted(list(bus.items())):
    print(f"Seat: {seat} \nDog: {dog}\n")

if(len(bus) < MAX_SEATS):
    available_seats = list(set(bus_seats).symmetric_difference(set(bus.keys())))

    print("WAHEY! There's room for one more!")
    print(f"Available seats are: {available_seats}")
    pet_name = input("Enter pet name: ")
    pet_breed = input("Enter pet breed: ")
    pet_pickup_time = input("Enter pickup time: ")
    pet_droppoff_time = input("Enter dropoff time: ")

    print(str(int(sorted(bus.keys())[-1])+1))
    bus[available_seats[0]] = {'name': pet_name, 'breed': pet_breed, 'pickup_time': pet_pickup_time, 'dropoff_time': pet_droppoff_time}


while True:
    pet_leaving = input("Which pet is leaving?")
    if(pet_leaving == ""):
        break
    
    found_pet = False
    for seat, pet in bus.items():
        if(pet["name"] == pet_leaving):
            del bus[seat] #Can use bus.pop(pet_leaving) too. Returns the popped value
            found_pet = True
            break
    
    if(found_pet == False):
       print("Could not find pet, pls enter again")
    else:
        print(f"{pet_leaving} from seat {seat} is leaving off early and heading home!")
        break
```

```python
#Quick problem I wanted to solve:
#Find all value pairs/solutions under n where a^3 + b^3 = c^3 + d^3

#BIIG NOTE: default dict creates non-existent keys automatically
from collections import defaultdict


def return_pairs(n):
    values_dict = defaultdict(list) #Initializes a dict with a key : list structure

    #pre-compute all the cubes to save runtime and overhead
    cubes = [i**3 for i in range(n+1)]

    for a in range(1, n+1):
        for b in range(a, n+1): #Start from a to avoid duplicate pairs
            result = cubes[a] + cubes[b]
            values_dict[result].append((a,b))
    
    solutions = {k : v for k, v in values_dict.items() if len(v) > 1}
    return solutions

out = return_pairs(2000)
print(out)
```

```python
#Quick problem I wanted to solve:
#Find all value pairs/solutions under n where a^3 + b^3 = c^3 + d^3

#BIIG NOTE: default dict creates non-existent keys automatically
from collections import defaultdict


def return_pairs_no_pre_computed_cubes(n):
    values_dict = defaultdict(list) #Initializes a dict with a key : list structure

    for a in range(1, n+1):
        for b in range(a, n+1): #Start from a to avoid duplicate pairs
            result = a**3 + b**3
            values_dict[result].append((a,b))
    
    solutions = {k : v for k, v in values_dict.items() if len(v) > 1}
    return solutions

def return_pairs_all_optimizations(n):
    values_dict = defaultdict(list) #Initializes a dict with a key : list structure

    #pre-compute all the cubes to save runtime and overhead
    cubes = [i**3 for i in range(n+1)]

    for a in range(1, n+1):
        for b in range(a, n+1): #Start from a to avoid duplicate pairs
            result = cubes[a] + cubes[b]
            values_dict[result].append((a,b))
    
    solutions = {k : v for k, v in values_dict.items() if len(v) > 1}
    return solutions

def return_pairs_repeated_pairs(n):
    values_dict = defaultdict(list) #Initializes a dict with a key : list structure

    #pre-compute all the cubes to save runtime and overhead
    cubes = [i**3 for i in range(n+1)]

    for a in range(1, n+1):
        for b in range(1, n+1): #Start from a to avoid duplicate pairs
            result = cubes[a] + cubes[b]
            values_dict[result].append((a,b))
    
    solutions = {k : v for k, v in values_dict.items() if len(v) > 1}
    return solutions

n=1000
#Note that the "number" parameter refers to how many times each test is run. The average is then returned to get the algorithm's time performance
print(timeit.timeit('return_pairs_no_pre_computed_cubes(n)', globals=globals(), number=10))
print(timeit.timeit('return_pairs_all_optimizations(n)', globals=globals(), number=10))
print(timeit.timeit('return_pairs_repeated_pairs(n)', globals=globals(), number=10))
```

```python
def encode(message, shift):
    result = []
    for c in message:
        if c.isalpha():
            base = ord('A') if c.isupper() else ord('a')
            result.append(chr((ord(c) - base + shift) % 26 + base))
        else:
            result.append(c)
    return "".join(result)


def decode(encoded_message, shift):
    return encode(encoded_message, -shift)  # decoding is just encoding backwards


# Example
shift = 3
message = "Super Secret Message"
encoded_message = encode(message, shift)
decoded_message = decode(encoded_message, shift)

print(f"Original: {message}")
print(f"Encoded : {encoded_message}")
print(f"Decoded : {decoded_message}")
```

```python
import string

keys = string.ascii_lowercase + string.ascii_uppercase + string.digits + string.whitespace + string.punctuation
print(keys)
shuffled_keys = "".join(random.sample(keys, len(keys)))

encode_dict = {key : value for (key, value) in zip(keys, shuffled_keys)}
decode_dict = {key : value for (key, value) in zip(shuffled_keys, keys)}

#print(encode_dict)
#print(decode_dict)

def decode(message):
    return "".join(decode_dict[c] for c in message)

def encode(message):
    return "".join(encode_dict[c] for c in message)

message = "This is a message"
encoded_message = encode(message)
decoded_message = decode(encoded_message)
print(f"This is the original message: {message}")
print(f"This is the encoded message: {encoded_message}")
print(f"This is the decoded message: {decoded_message}")
```

```python
print("Math Tutor")

import random
import timeit

#return tuple: result, problem string
def generate_question(max_range):
    a = random.randrange(-max_range, max_range)
    b = random.randrange(-max_range, max_range)
    return a*b , f"{a}x{b}"

def game():
    while True:
        try:
            question_amount = int(input("Enter amount of questions you'd like: "))
        except:
            print("Invalid input")
        else:
            break
    
    #Initialize dictionary
    solutions_dict = {key : [] for key in range(question_amount)}

    correct_questions_amount = 0
    for i in range (question_amount):
        solution, problem = generate_question(10)
        
        solutions_dict[i].append(problem)
        solutions_dict[i].append(solution)

        while True:
            try:
                user_solution = int(input(f"What's {problem}: "))
            except:
                print("Invalid input")
            else:
                break
        
        if(user_solution == solution):
            print("Correct!")
            correct_questions_amount += 1
        else:
            print("No dice!")

    print("Thank you for playing!")
    print(f"You've got {correct_questions_amount} out of {question_amount}")
    print(f"Your score is {(correct_questions_amount/question_amount)*100}%")
    
    print()
    print("=====SOLUTIONS=====")
    for question, solution in solutions_dict.items():
        print(f"{question + 1} : {solution[0]} = {solution[1]}")

    return 1 #Remember to return someting for the timeit library

time_to_complete = round(float(timeit.timeit('game()', globals=globals(), number=1)), 2)

print(f"You took {time_to_complete//60} minutes and {time_to_complete%60} seconds")
```

```python
#Marble game
import random

#Initialize the marble bag
marble_bag = ["green"]*6 + ["red"]*4
random.shuffle(marble_bag)
print(marble_bag)

coins = 1000

while True:
    try:
        n_rounds = int(input("Enter how many rounds you wish to play: "))

    except:
        print("Invalid Input")
    
    else:
        break

for i in range(n_rounds):
    while True:
        try:
            user_bet = int(input(f"You have {coins}. How much do you want to bet? "))

        except:
            print("Invalid Input")
        
        else:
            break
    
    coins -= user_bet
    marble = random.choice(marble_bag)

    print(f"You drew {marble}")
    if(marble == "green"):
        print(f"You win {user_bet} coins back")
        coins += 2*user_bet
    
    if(marble == "red"):
        print(f"You loose {user_bet} coins back")

    print(f"Your coin count is {coins}")
    print()

    if(coins <= 500):
        print(f"Your coin count is below 500")
        print("Game Over")
        break
```
